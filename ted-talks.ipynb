```python
ted_data = pd.read_csv('../input/ted-talks/data.csv')
# Data Overview 
ted_data.head()
```

```python
# Dropping the missing author name column
ted_data.dropna(inplace=True)
ted_data.isnull().any().sum()
```

```python
# Data Distribution and correlation
plt.figure(figsize=(200,200))
sns.pairplot(ted_data);
```

### As the is a very strong correlation between likes and views, we need a variable to rely on

```python
# Creating like ratio column
['like_ratio'] = ted_data.likes/ted_data.views
```

```python
sns.boxplot(x= 'like_ratio', data= ted_data)
```

```python
ted_data.describe()
```

```python
# # Author who has the best like ratio  
ted_data.query('like_ratio == like_ratio.max()')[['author','like_ratio', 'title', 'link']]
```

### Consider watching this Talk

```python
# Creating variable to group data by author
author = ted_data.groupby('author')
```

```python
author_state = author.agg({'link':'count',
                               'likes':'mean',
                               'views':'mean',
                               'like_ratio':'mean'
                              }).sort_values(by='like_ratio', ascending=False )
author_state.head(25)
```

### Top 25 authors with the like ratio 
###### However, they all have only one Ted-talk, and about half have relatively low views

# let's have a look at the authors with a high view rate 
#### considering 

```python
author_state = author.agg({'link':'count',
                            'views':'sum',
                            'like_ratio':'mean'
                            })
author_df = author_state.reset_index()
author_df.describe()
```

```python
sns.boxplot(x ='like_ratio', data= author_df) ;
```

```python
# From my point of view, I will filter the data to contain only how is above the median for both views and like ratio 
# You can change the percentage to be whatever you want and sort the data accordingly 
filter_data = author_df[(author_df.views >= author_df.views.quantile(0.5)) & (author_df.like_ratio >= author_df.like_ratio.quantile(0.5))]

filter_data.sort_values(by=['views','like_ratio'], ascending=False).head(25)
```
